```python
import numpy as np
import matplotlib.pyplot as plt

# Lista original de columnas (las que tiene tu DataFrame)
years = ['AVG 2020', 'AVG 2021', 'AVG 2022', 'AVG 2023', 'AVG 2024']

# Etiquetas que querés mostrar en la leyenda y el gráfico
year_labels = ['2020', '2021', '2022', '2023', '2024']

labels = df['GEO (Labels)']
x = np.arange(len(labels))
width = 0.15

# Crear figura
plt.figure(figsize=(14, 7))

# Dibujar barras para cada año, usando etiquetas sin "AVG"
bars = []
for i, year in enumerate(years):
    bars.append(plt.bar(x + i*width, df[year], width))  # Dibujar barras y guardarlas

# Calcular mediana global
median_value = df[years].values.flatten()
median = np.nanmedian(median_value)

print(f"Mediana global: {median}")  # 👉 Esto te lo imprimirá en la consola de Colab

# Línea horizontal de mediana
plt.axhline(y=median, color='crimson', linestyle='--', linewidth=2.5, label=f'Mediana global: {median:.2f}')

# Estética
plt.xticks(x + width*2, labels, rotation=90)
plt.ylabel('Tasa de Empleabilidad (%)')
plt.title('Tasa de Empleabilidad por País (2020–2024)')

# Crear una nueva leyenda con los años limpios (sin 'AVG')
handles = [bar[0] for bar in bars]  # Se obtiene el primer objeto de cada conjunto de barras
plt.legend(handles=handles, labels=year_labels, title="Año")  # Usamos los year_labels para la leyenda

plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
```

#  Clasificación de países en grupos según su tasa de empleabilidad (Modelo de Clasificación)


#K Means


```python
# TODOS LOS GENEROS
import pandas as pd

# Cargar solo una hoja del Excel
df = pd.read_excel("comparativa.xlsx", sheet_name="total")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df.dtypes
```

```python
# analizo solo 2024
X = df[['AVG 2024']].values
```

```python
# Calcular inercia para diferentes valores de k
inertia = []
K_range = range(1, 6)  # Pruebas de 1 a 5 clusters
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
# Graficar el método del codo
plt.figure(figsize=(8, 4))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo para elegir k óptimo')
plt.xticks(K_range)
plt.grid()
plt.show()
```

```python
# Aplico K-Means con 2 klusters
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X)  # Añade la columna de clusters al DataFrame
```

```python
# visualización cluster
plt.scatter(df["GEO (Labels)"], df["AVG 2024"], c=df["Cluster"], cmap="viridis")
plt.xticks(rotation=90)
plt.xlabel("Países")
plt.ylabel("Empleabilidad 2024 (%)")
plt.title("Clusters de empleabilidad (k=2)")
plt.colorbar(label="Cluster")
plt.show()
```

```python
import pandas as pd

# Cargar solo una hoja del Excel
df = pd.read_excel("historic_data.xlsx", sheet_name="female 45-49")

df = df.drop([0, 1, 2])
# Ver los primeros datos
df.head()
```

```python
df.dtypes
```

```python
#crear una columna nueva con el average de los ultimos 10 años

# Lista de columnas de años
years_last_10 = list(range(2015, 2025))

# Creo la columna con la media (ignorando NA por defecto), redondeo a dos decimales
df['avg_last_10_year'] = df[years_last_10].mean(axis=1).round(2)
```

```python
print(df.index)
```

```python
# Graficar

import matplotlib.pyplot as plt
import numpy as np

# Ordenar el DataFrame por el promedio (opcional, para una mejor visualización)
df_sorted = df.sort_values('avg_last_10_year', ascending=False)

# Datos del gráfico
labels = df_sorted['TIME']
values = df_sorted['avg_last_10_year']
x = np.arange(len(labels))

# Calcular mediana global
median = np.nanmedian(values)

# Crear la figura
plt.figure(figsize=(14, 7))
bars = plt.bar(x, values, color='skyblue', edgecolor='black')

# Mostrar el valor encima de cada barra
for i, value in enumerate(values):
    plt.text(
        x=i,
        y=value + 0.5,  # ligeramente arriba
        s=f'{value:.1f}%',
        ha='center',
        va='bottom',
        fontsize=9,
        color='black'
    )

# Línea de mediana
plt.axhline(median, color='crimson', linestyle='--', linewidth=2, label=f'Mediana: {median:.2f}')

# Personalización
plt.xticks(x, labels, rotation=90)
plt.ylabel('Promedio de Tasa de Empleabilidad (%)')
plt.title('Empleabilidad Media Mujeres Edad 45-49 (2015–2024) por País')
plt.legend()
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Comparativa por género:

```python
df_male = pd.read_excel("historic_data.xlsx", sheet_name="males 45-49")
df_female = pd.read_excel("historic_data.xlsx", sheet_name="female 45-49")

df_female = df.drop(df.columns[0], axis=1)
df_male = df_male.drop([0, 1, 2], axis=0).reset_index(drop=True)

df_male['avg_last_10_year'] = df_male[years_last_10].mean(axis=1).round(2)
df_female['avg_last_10_year'] = df_female[years_last_10].mean(axis=1).round(2)

df_male.head()
# df_female.head()
```

```python
df_male.dtypes
```

```python
# Solo quedarnos con las columnas necesarias y renombrarlas
df_male = df_male[['TIME', 'avg_last_10_year']].rename(columns={'avg_last_10_year': 'male_avg'})
df_female = df_female[['TIME', 'avg_last_10_year']].rename(columns={'avg_last_10_year': 'female_avg'})

# Unir por 'TIME' (países)
df_combined = pd.merge(df_male, df_female, on='TIME')

# Ordenar por promedio femenino)
df_combined = df_combined.sort_values('female_avg', ascending=False)

# Etiquetas y posiciones de las barras
labels = df_combined['TIME']
x = np.arange(len(labels))
width = 0.35  # ancho de las barras

# Crear la figura, ajustando el tamaño para dar más espacio
plt.figure(figsize=(15, 9))

# Crear barras para hombres y mujeres
bars_male = plt.bar(x - width/2, df_combined['male_avg'], width, label='Hombres', color='steelblue')
bars_female = plt.bar(x + width/2, df_combined['female_avg'], width, label='Mujeres', color='lightcoral')

# Calcular la mediana global y dibujarla
combined_values = pd.concat([df_combined['male_avg'], df_combined['female_avg']])
median_global = np.nanmedian(combined_values)
plt.axhline(median_global, color='gray', linestyle='--', linewidth=2, label=f'Mediana Global: {median_global:.2f}%')

# Personalización del gráfico
plt.xticks(x, labels, rotation=90)
plt.ylabel('Empleabilidad Promedio (%)')
plt.title('Empleabilidad Media por Género (Edad 45–49) [2015–2024]')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Ajustamos la leyenda fuera del gráfico

# Cambiar el intervalo del eje Y para mostrar de 5 en 5
plt.yticks(np.arange(0, max(df_combined[['male_avg', 'female_avg']].max()) + 5, 5))

# Añadir líneas de la cuadrícula
plt.grid(axis='y', linestyle=':', alpha=0.5)

# Ajustar el diseño para mayor espacio
plt.tight_layout()
plt.show()
```
